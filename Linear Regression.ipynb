# Simple Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
print(df.head())
```

```python
X = df[['YearsExperience']]
y = df['Salary']
plt.scatter(X,y)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, y)
```

```python
lr.predict([[2]])
```

### Find Coefficient

```python
slope=lr.coef_
```

### Find Intercept

```python
intercept=lr.intercept_
```

### Make Predictions

```python
# Make predictions
line=X*slope+intercept
X = df[['YearsExperience']]
y = df['Salary']
plt.scatter(X,y)
plt.plot(X,line,color='red')
```
